```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Iris dataset
df = pd.read_csv("iris.csv")
```

```python

# --- 1. Features and their types ---
print("Features and Data Types:")
print(df.dtypes)
print("\nShape:", df.shape)
print("Missing values:", df.isnull().sum().sum())
```

```python
# --- 2. Histograms for each numeric feature ---
df.hist(figsize=(10,8), bins=15, color='steelblue', edgecolor='black')
plt.suptitle("Feature Distributions - Iris Dataset")
plt.tight_layout()
plt.show()

# Individual histogram
sns.histplot(df['sepal_length'], bins=15, kde=True)
plt.title("Sepal Length Distribution")
plt.show()
```

```python

# --- 3. Box plots for each feature ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    ax = axes[i//2, i%2]
    sns.boxplot(x='species', y=feature, data=df, ax=ax)
    ax.set_title(f"{feature} by Species")

plt.tight_layout()
plt.show()
```

```python

# --- 4. Identify Outliers using IQR ---
for col in features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1-1.5*IQR) | (df[col] > Q3+1.5*IQR)]
    print(f"{col}: {len(outliers)} outliers")
```
