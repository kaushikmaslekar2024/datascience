```python
# ---- PART 2: Iris dataset statistics by species ----
df = pd.read_csv('iris.csv')

# Basic describe for all species
print(df.describe())
```

```python
# Per species statistics
for species in df['species'].unique():#
    print(f"\n--- {species} ---")
    print(df[df['species']==species].describe())
```

```python
# Percentile calculation
print("\n25th percentile:", df['sepal_length'].quantile(0.25))
print("50th percentile:", df['sepal_length'].quantile(0.50))
print("75th percentile:", df['sepal_length'].quantile(0.75))
print("Mean:", df['sepal_length'].mean())
print("Std:", df['sepal_length'].std())
```

```python
# Manual variance and std
vals = df['sepal_length']
mean = vals.mean()
variance = ((vals - mean)**2).sum() / (len(vals) - 1)
std = variance**0.5
print("Manual Variance:", variance)
print("Manual Std Dev:", std)
```
