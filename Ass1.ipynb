#Data Wrangling I

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
# 1. Load dataset
df = pd.read_csv('iris.csv')
```

```python
# 2. Basic info
print(df.head())
print(df.shape)          # dimensions
print(df.dtypes)         # data types
print(df.isnull().sum()) # missing values
print(df.describe())     # statistics
```

```python
# 3. Data Formatting - change data type
df['petal_length'] = df['petal_length'].astype("int")
print(df.dtypes)
```

```python
# 4. Data Normalization (Min-Max Scaling)
min_max_scaler = preprocessing.MinMaxScaler()
x = df.iloc[:, :4]        # only numeric columns
x_scaled = min_max_scaler.fit_transform(x)
df_normalized = pd.DataFrame(x_scaled, columns=df.columns[:4])
print(df_normalized.head())
```

```python
# 5. Label Encoding (categorical to numeric)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['species_encoded'] = le.fit_transform(df['species'])
print(df[['species','species_encoded']].head())
```

```python
# 6. One-Hot Encoding
df_onehot = pd.get_dummies(df, columns=['species'])
print(df_onehot.head())
```
