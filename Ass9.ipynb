Data Visualization II

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic
df = sns.load_dataset('titanic')
```

```python
# --- Box Plot: Age by Gender ---
plt.figure(figsize=(8,5))
sns.boxplot(x='sex', y='age', data=df)
plt.title("Age Distribution by Gender")
plt.show()
```

```python
# --- Box Plot: Age by Gender + Survived (hue) ---
plt.figure(figsize=(10,6))
sns.boxplot(x='sex', y='age', data=df, hue='survived')
plt.title("Age by Gender and Survival Status")
plt.legend(title='Survived', labels=['No','Yes'])
plt.show()
```

```python
# --- Understanding the BoxPlot output ---
# For each gender:
print("Male age stats:")
print(df[df['sex']=='male']['age'].describe())
print("\nFemale age stats:")
print(df[df['sex']=='female']['age'].describe())
```

```python

# --- Additional: Violin Plot comparison ---
plt.figure(figsize=(10,6))
sns.violinplot(x='sex', y='age', data=df,
               hue='survived', split=True)
plt.title("Age Distribution: Gender vs Survival")
plt.show()
```
