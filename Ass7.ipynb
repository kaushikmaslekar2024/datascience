Text Analytics

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
# Sample document
text = "Data science is an interdisciplinary field. Data scientists use scientific methods."
```

```python
# 1. TOKENIZATION
words = word_tokenize(text)
sentences = sent_tokenize(text)
print("Word tokens:", words)
print("Sentences:", sentences)
```

```python
# 2. STOP WORDS REMOVAL
stop_words = set(stopwords.words('english'))
filtered = [w for w in words if w.lower() not in stop_words and w.isalnum()]
print("After stopword removal:", filtered)
```

```python
# 3. STEMMING
ps = PorterStemmer()
stemmed = [ps.stem(w) for w in filtered]
print("Stemmed:", stemmed)
```

```python
# 4. LEMMATIZATION
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w) for w in filtered]
print("Lemmatized:", lemmatized)
```

```python
# 5. POS TAGGING
pos_tags = pos_tag(words)
print("POS Tags:", pos_tags)
```

```python
# 6. TF-IDF
docA = "Jupiter is the largest planet"
docB = "Mars is the fourth planet from the sun"
corpus = [docA, docB]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out(),
    index=['DocA','DocB']
)
print("\nTF-IDF Matrix:\n", df_tfidf)
```
