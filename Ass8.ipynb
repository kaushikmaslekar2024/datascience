Data Visualization I

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# Load Titanic dataset (inbuilt in Seaborn)
df = pd.read_csv('titanic.csv')
print(df.head())
print(df.shape)      # 891 rows, 15 columns
print(df.columns.tolist())
```

```python

# --- 1. HISTOGRAM of Fare ---
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python

# --- 2. Count Plot - how many survived ---
sns.countplot(x='2urvived', data=df)
plt.title("Survival Count")
plt.show()
```

```python

# --- 3. Bar Plot - avg fare by class ---
sns.barplot(x='Pclass', y='Fare', data=df)
plt.title("Average Fare by Class")
plt.show()
```

```python

# --- 4. Dist Plot (distribution) ---
sns.histplot(df['Age'].dropna(), kde=True, bins=20)
plt.title("Age Distribution")
plt.show()
```

```python

# --- 5. Heat Map (correlation) ---
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python

# --- 6. Violin Plot ---
sns.violinplot(x='Sex', y='Age', data=df)
plt.title("Age by Gender")
plt.show()
```
